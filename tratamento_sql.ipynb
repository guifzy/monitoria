```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\guipi\Desktop\python\api oficina\Oficina_Sentment140\oficina\Oficina Banco de Dados.csv')
df.columns
```

```python
df = df[['Nome completo', 'RA', 'E-mail (@sempreceub)']]
```

```python
df['Nome completo'] = df['Nome completo'].str.lower()
df['E-mail (@sempreceub)'] = df['E-mail (@sempreceub)'].str.lower()
df = df.sort_values(by="Nome completo")
df.reset_index(inplace=True)
df.drop(columns='index', inplace=True)
df
```

```python
atualizacoes = {
    'eduardo borges lemos': 'eduardoborgeslemos@sempreceub.com',
    'eduardo sebben rocha': 'eduardo.srocha@sempreceub.com',
    'josé paulo nascimento lupatini': 'jplupatini07@sempreceub.com'
}

for nome, email in atualizacoes.items():
    df.loc[df['Nome completo'] == nome, 'E-mail (@sempreceub)'] = email
```
